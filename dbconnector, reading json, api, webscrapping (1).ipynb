```python
import requests
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/List_of_prime_ministers_of_India"
```

```python
r=requests.get(url)
```

```python
htmlcontent=r.content
```

```python
htmlcontent
```

```python
soup=BeautifulSoup(htmlcontent,'html.parser')
```

```python
in_table = soup.find('table',{"class":"wikitable"})
in_table
```

```python
df = pd.read_html(str(in_table))
```

```python
df = pd.DataFrame(df[0])
df.columns=["No","No.1","Portrait","Name", "Party" ,"Took office","Left office","Time in office","Lok Sabha[e]","Ministry","Appointed by","Unnamed: 11_level"]
df
```

```python
df[["Name","Party","Took office","Left office"]]
```

```python
#
df["Took office"]=df["Took office"].str.replace("[§]","")
df["Took office"]=df["Took office"].str.replace("[[]]","")
df["Left office"]=df["Left office"].str.replace("[†]","")
df["Left office"]=df["Left office"].str.replace("[NC]","")
df["Left office"]=df["Left office"].str.replace("[RES]","")
df["Left office"]=df["Left office"].str.replace("[[]]","")
```

```python
df[["Name","Party","Took office","Left office"]]
```

```python
df1 = df.drop(26)
df1[["Name","Party","Took office","Left office"]]
```

```python
df1['Left office'].mask(df1['Left office'] == '10 ovember 1990', '10 November 1990', inplace=True)
df1
```

```python
df1['Left office']=pd.to_datetime(df1['Left office'],infer_datetime_format=True)
df1
```

```python
df1['Took office']=pd.to_datetime(df1['Took office'],infer_datetime_format=True)
df1
```

```python
df4=df1[["Name","Party","Took office","Left office"]]
df4.columns=["Name","Party","Took_office","Left_office"]
df4
```

```python
from sqlalchemy import create_engine
```

```python
df2=create_engine("mysql+pymysql://root:Brobrobrobro9@localhost/ministers")
```

```python
df4.to_sql("primeministers", df2 ,if_exists ='replace',index=False)
```
